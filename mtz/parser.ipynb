```python
from bs4 import BeautifulSoup
import requests

r  = requests.get("http://mtzmedservice.com/")

data = r.text
soup = BeautifulSoup(data, 'html.parser')
```

```python
import re
```

```python
for el in soup.find_all('ul','ex_menu')[0].find_all('li'):
    print( '<li>' + re.sub( "[^а-яА-Я-()]+"," ", el.text ) + '</li>' )
```
